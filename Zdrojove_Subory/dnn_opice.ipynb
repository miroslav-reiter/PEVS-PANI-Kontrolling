```python
import os

def load_images(path):
    ''' Put files into lists and return them as one list with all images
     in the folder'''
    image_files = sorted([os.path.join(path, file)
                          for file in os.listdir(path)
                          if file.endswith(".jpeg") or file.endswith(".jpg") or file.endswith(".png")])
    return image_files

n0_obrazky = load_images("data/train/n0")
n1_obrazky = load_images("data/train/n1")
n2_obrazky = load_images("data/train/n2")
n3_obrazky = load_images("data/train/n3")
n4_obrazky = load_images("data/train/n4")
n5_obrazky = load_images("data/train/n5")
n6_obrazky = load_images("data/train/n6")
n7_obrazky = load_images("data/train/n7")
n8_obrazky = load_images("data/train/n8")
n9_obrazky = load_images("data/train/n9")

img=plt.imread(n0_obrazky[0])
print(f"Rozmer obrazka ({n0_obrazky[0]}) je: {img.shape} px")

plt.title(n0_obrazky[0])
plt.imshow(plt.imread(n0_obrazky[0]))
plt.show()
```

## Klasifikácia obrázkov, využitie ImageDataGeneratora z TensorFlow

```python
# Prevzaté z Kaggle (https://www.kaggle.com/code/gaber0512/monkey-species-convolutional-neural-network) a
# TensorFlow (https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image/ImageDataGenerator)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
from pathlib import Path

train = Path('data/train/')
test = Path('data/validation/')

train_datagen = ImageDataGenerator(
        rescale = 1./255,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(
        train,
        target_size = (128, 128),
        batch_size = 32,
        class_mode = 'categorical')
validation_generator = test_datagen.flow_from_directory(
        test,
        target_size = (128, 128),
        batch_size = 32,
        class_mode = 'categorical')
```

```python
train = Path('data/train/')
test = Path('data/validation/')

train_filepaths = list(train.glob(r'**/*.jpg'))
test_filepaths = list(test.glob(r'**/*.jpg'))
```

## Štruktúra modelu

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Flatten,Dropout

model = Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(128,128,3), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(128,128,3), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(150, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer ='adam', metrics= ['accuracy'])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',patience=5)
```

## Model Fit

```python
model.fit_generator(train_generator, epochs = 20, validation_data=validation_generator, callbacks=early_stop, verbose = 1)
```

```python
loses = pd.DataFrame(model.history.history)
loses.head()
```

## Predikcia modelu

```python
def proc_img(filepath):
    """
    Vytvori DataFrame s cestou k súboru a menovkami obrázkov
    :param filepath:
    :return: DataFrame
    """
    labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepath))

    filepath = pd.Series(filepath, name='Filepath').astype(str)
    labels = pd.Series(labels, name='Label')

    df = pd.concat([filepath, labels], axis=1)

    # Premieša DataFrame a resetuje index
    df = df.sample(frac=1).reset_index(drop = True)

    return df
```

```python
train_df = proc_img(train_filepaths)
test_df = proc_img(test_filepaths)
```

```python
pred_labels = model.predict(validation_generator)
pred_labels = np.argmax(pred_labels,axis=1)

labels = (validation_generator.class_indices)
labels = dict((v,k) for k,v in labels.items())
pred_labels = [labels[k] for k in pred_labels]
```

```python
def pred_imag(index):
    """
    Testuje predikciu obrazkov
    :param index: obrazka
    :return:
    """
    plt.imshow(plt.imread(test_df.Filepath.iloc[index]))
    if test_df.Label.iloc[index]==pred_labels[index]:
        plt.title(f" True prediction model {test_df.Label.iloc[index]} , {pred_labels[index]}")
    else:
        plt.title(f" False prediction model {test_df.Label.iloc[index]} , {pred_labels[index]}")

    plt.show()
```
